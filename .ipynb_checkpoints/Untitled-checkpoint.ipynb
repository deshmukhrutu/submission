```python
def get_type_count(df)->dict:
    df['car_type']=pd.cut(df['car'],bins=[-float('inf'),15,25,float('inf')],labels=['low','medium','high'],right=False)
    type_count=df['car_type'].value_counts().to_dict()
    sorted_type_counts=dict(sorted(type_count.items()))
    return sorted_type_counts
```

```python
def get_bus_indexes(df)->list:
    mean_bus_value=df['bus'].mean()
    bus_indexes=df[df['bus']>2*mean_bus_value].index.tolist()
    bus_indexes.sort()
    return bus_indexes
```

```python
result=get_bus_indexes(df)
```

```python
def filter_routes(df) -> list:    
    route_avg_truck=df.groupby('route')['truck'].mean()
    filtered_routes=route_avg_truck[route_avg_truck>7].index.tolist()
    filtered_routes.sort()

    return filtered_routes
```

```python
def time_check(df)->pd.Series:    
    
    df['start_timestamp']=pd.to_datetime(df['startDay'] + ' ' + df['startTime'])
    df['end_timestamp'] = pd.to_datetime(df['endDay'] + ' ' + df['endTime'])
    incomplete_time_mask=(
        (df['start_timestamp'].dt.time != pd.to_datetime('00:00:00').time()) |
        (df['end_timestamp'].dt.time != pd.to_datetime('23:59:59').time())   |
        (df['start_timestamp'].dt.dayofweek > df['end_timestamp'].dt.dayofweek) |
        ((df['end_timestamp']-df['start_timestamp']).dt.total_seconds()>864000)
    )
    result_series=df.groupby(['id','id_2']).any()[incomplete_time_mask]
    return result_series
```
